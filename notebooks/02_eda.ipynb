# Exploratory Data Analysis (EDA)

This notebook explores sales trends, product performance, deal sizes,
and regional patterns to understand how the business is performing.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_excel(r"D:\549\sales_data\sales_data_cleaned.xlsx")
```

### Key Business Questions Explored

- How are sales distributed over time?
- Which product lines generate the most revenue?
- Which countries contribute the highest sales?
- How are deal sizes distributed?

```python
sales_by_year = df.groupby("YEAR_ID")["SALES"].sum()

plt.figure(figsize=(4,2))
sales_by_year.plot(kind="bar")
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

**Observation:** Sales vary across years, indicating changes in demand and business performance over time.

```python
plt.figure(figsize=(10,5))
df.groupby("PRODUCTLINE")["SALES"].sum().sort_values().plot(kind="barh")
plt.title("Sales by Product Line")
plt.xlabel("Sales") 
plt.ylabel("Product Line") 
plt.show()
```

**Observation:** A small number of product lines contribute a significant portion of total revenue.

```python
plt.figure(figsize=(10,5))
df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False).head(10).plot(kind="bar") 
plt.title("Top 10 Countries by Sales") 
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()
```

**Observation:** Sales are concentrated in a few key countries, suggesting strong regional markets.

```python
sns.countplot(x="DEALSIZE", data=df)
plt.title("Deal Size Distribution") 
plt.show()
```

### EDA Summary

- Sales performance differs across years.
- Certain product lines dominate revenue contribution.
- Revenue is concentrated in a limited number of countries.
- Medium and Large deals form the majority of transactions.
