# Business Insights & Recommendations

This notebook summarizes key insights derived from the sales data
and translates them into actionable business recommendations.

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\549\sales_data\sales_data_cleaned.xlsx")
```

### Key Performance Indicators (KPIs)

```python
total_sales = df["SALES"].sum()
average_order_value = df["SALES"].mean()
total_orders = df["ORDERNUMBER"].nunique()

total_sales, average_order_value, total_orders
```

```python
total_sales = df["SALES"].sum()
total_orders = df["ORDERNUMBER"].nunique()

aov = total_sales / total_orders

aov
```

**Interpretation:**

- Total Sales represent overall revenue generated.
- Average Order Value indicates customer purchase behavior.
- Total Orders show transaction volume.

```python
df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
```

```python
df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False).head(5)
```

### Key Business Insights

1. A small set of product lines drives the majority of revenue.
2. Sales are heavily concentrated in a few countries.
3. Medium and Large deals contribute most to total sales.
4. Average order value suggests potential for upselling strategies.

### Business Recommendations

- Focus marketing and inventory planning on top-performing product lines.
- Strengthen operations in high-revenue countries.
- Introduce targeted promotions to convert Small deals into Medium deals.
- Use historical sales trends to support demand forecasting.

### Why These Insights Matter

These insights help decision-makers prioritize products, regions,
and deal strategies to maximize revenue and operational efficiency.
